```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
with open("./logs_cnn_1a-iter200.txt") as f:
    s = f.read()
```

```python
trains = [s.split("\n")[i+1] for i in range(len(s.split("\n"))) if "EP_train" in s.split("\n")[i]]
test  = [s.split("\n")[i+1] for i in range(len(s.split("\n"))) if "EP_test" in s.split("\n")[i]]
```

```python
trains_loss = [float(i.split("'avg_loss': ")[1].split(", 'l")[0]) for i in trains]
test_loss = [float(i.split("'avg_loss': ")[1].split(", 'l")[0]) for i in test]
```

```python
plt.plot(trains_loss)
plt.plot(test_loss)
```
