```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("loss_800.log")
```

```python
df.head()
```

```python
plt.plot(df['train_loss'], label="train")
plt.plot(df['test_loss'], label="test")
plt.legend(bbox_to_anchor=(1, 1), loc='upper right', borderaxespad=0, fontsize=10)
```

```python
plt.plot(df['train_F1_score'], label="train")
plt.plot(df['test_F1_score'], label="test")
plt.legend(bbox_to_anchor=(1, 1), loc='upper right', borderaxespad=0, fontsize=10)
```
